```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("3 RD PROJECT DHANYA26.csv")

# ---------------------------
# Calculate Average Marks
# ---------------------------
subjects = ["Java", "Python", "Biometrics"]

df["Average"] = df[subjects].mean(axis=1)

# ---------------------------
# Assign Grades
# ---------------------------
def assign_grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 50:
        return "D"
    else:
        return "F"

df["Grade"] = df["Average"].apply(assign_grade)

# ---------------------------
# Pass / Fail Status
# ---------------------------
def pass_fail(row):
    if all(row[sub] >= 40 for sub in subjects):
        return "Pass"
    else:
        return "Fail"

df["Status"] = df.apply(pass_fail, axis=1)

# Display result
print("\n3 RD PROJECT DHANYA26\n")
print(df[["Student Name", "Roll Number", "Java", "Python", "Biometrics"]])

# Save report
df.to_csv("3 RD PROJECT DHANYA26.csv", index=False)
display(df)
```

```python
subject_avg = df[subjects].mean()

plt.figure(figsize=(8,5))
subject_avg.plot(kind="bar")
plt.title("Subject-wise Average Marks")
plt.ylabel("Average Marks")
plt.xlabel("Subjects")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["Grade"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Grade Distribution")
plt.ylabel("")
plt.show()
```

```python
class_avg = df.groupby('Grade')['Average'].mean()

plt.figure(figsize=(6,4))
class_avg.plot(kind="bar")
plt.title("Class-wise Average Performance")
plt.ylabel("Average Marks")
plt.xlabel("Class")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Student Marks Data
data = {
    "Student Name": ["Dhanya", "Minu", "Akshaya", "Neena", "Gopika"],
    "Java": [89, 87, 67, 78, 66],
    "Python": [96, 87, 88, 80, 76],
    "Biometrics": [88, 78, 77, 80, 81]
}

# Create DataFrame
df = pd.DataFrame(data)

# Subject-wise Comparison Table
subject_comparison = pd.DataFrame({
    "Subject": ["Java", "Python", "Biometrics"],
    "Average Marks": [
        df["Java"].mean(),
        df["Python"].mean(),
        df["Biometrics"].mean()
    ],
    "Highest Marks": [
        df["Java"].max(),
        df["Python"].max(),
        df["Biometrics"].max()
    ],
    "Lowest Marks": [
        df["Java"].min(),
        df["Python"].min(),
        df["Biometrics"].min()
    ]
})

print("\nSUBJECT-WISE COMPARISON TABLE")
print(subject_comparison)
```
